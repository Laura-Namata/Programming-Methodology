```python
# Question 1
# Part 1
# Importing neccesary libraries and getting the data.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
summarize,
poly)
from matplotlib.pyplot import subplots
import seaborn as sns
from ISLP import confusion_table
cars =pd.read_csv("DATA.csv")
```

```python
# Making neccesary adjustments to make the data meaning ful.
cars['transmission'].replace(['Manual','Automatic','Semi-Auto'],[0,1,2],inplace=True)
cars['fuelType'].replace(['Petrol','Diesel','Hybrid'],[0,1,2],inplace=True)
```

```python
# Part 2
# Using SPLOM to find the 4 best variables.
sns.pairplot(cars)
# Given the Scatterplot matrix below: I recommend we look at how mileage, transmission, miles per gallon(mpg) and engine size affect the price of the car.
```

```python
# Performing a correlation analysis on the 4 variables.
corr_matrix = cars.corr()
sns.heatmap(corr_matrix, annot = True)
# This shows that price has a high correlation with mileage, transmission,mpg and engine size thats why they were selected other than the rest.
```

```python
# Part 3
# Linear regression
x = MS(['mileage','tax','transmission','mpg','engineSize','fuelType']).fit_transform(cars)
y = np.array(cars['price'].values.tolist())
#using statsmodels
x = sm.add_constant(x) # adding a constant
 
model = sm.OLS(y, x).fit()
predictions = model.predict(x) 
 
print_model = model.summary()
print(print_model)
```

```python
# COEFFIECIENTS AND THEIR IMPLICATIONS.
# 1. Intercept ( 2.444e+04): When all independent variables are zero, the estimated average 'price' of the car is  24,440.

# 2  mileage (-0.2348): A one-unit increase in the 'mileage' variable is associated with a decrease of 0.2348 units in 'price', holding all other variables constant.

# 3. transmission (1139.1328): A one-unit increase in 'transmission' results in an increase of 1139.1328 units in 'price', holding all other variables constant.

# 4. tax (-24.4358): A one-unit increase in 'tax' leads to a decrease of 24.4358 units in 'price', holding all other variables constant.

# 5. engine size (10,140): A one-unit increase in 'engine size' results in an increase of 10,140 units in 'price', holding all other variables constant.

# 6. mpg (-273.6689): A one-unit increase in 'mpg' leads to a decrease of 273.6689 units in 'price', holding all other variables constant.

# 7. fuel type (1183.1228): A one-unit increase in 'fuel type' results in an increase of 1183.1228 units in 'price', holding all other variables constant.

# The model has an R-squared value of o.729, which means that 72.9% of the variance in the dependent variable is explained by the independent variables. The adjusted R-squared is 0.729, which accounts for the number of predictors in the model
# All of the coefficients are statistically significant with p-values less than 0.05, suggesting that they have a significant impact on the dependent variable.
```
